```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import cv2

from skindetection import SkinColorDetector
from measure import MeasureSkin
from preprocessing import CelebA
import mydata
```

```python
db = "CelebA"
metafile = "list_attr_celeba.csv"
root = f"../../dataset/AAAI 2025/{db}/"
df = pd.read_csv(root + metafile)
```

```python
ca = CelebA(root, df)
ca.unify_dataset()
ca.check_balance()
df_balanced = ca.balance_dataset()
df_train, df_valid, df_test = ca.split_dataset(df_balanced)
```

```python
non_frontal_index = ca.check_non_frontal_face(df_train)
df_train.drop(index = non_frontal_index, inplace=True)
df_train.groupby(["labels", "skin tone"]).size()
```

```python
non_frontal_index = ca.check_non_frontal_face(df_valid)
df_valid.drop(index = non_frontal_index, inplace=True)
df_valid.groupby(["labels", "skin tone"]).size()
```

```python
non_frontal_index = ca.check_non_frontal_face(df_test)
df_test.drop(index = non_frontal_index, inplace=True)
df_test.groupby(["labels", "skin tone"]).size()
```

```python
save_directory_path = f"{root}mask/"
df_new_train = ca.create_masked_image(df_train, save_directory_path)
print(f"Num of train dataset: {len(df_new_train)}")
```

```python
df_new_valid = ca.create_masked_image(df_valid, save_directory_path)
print(f"Num of valid dataset: {len(df_new_valid)}")
```

```python
df_new_test = ca.create_masked_image(df_test, save_directory_path)
print(f"Num of test dataset: {len(df_new_test)}")
```

```python
non_skin_index = ca.check_no_skin_indetified(df_new_train)
df_train.drop(index = non_skin_index, inplace=True)
df_train.groupby(["labels", "skin tone"]).size()
```

```python
non_skin_index = ca.check_no_skin_indetified(df_new_valid)
df_valid.drop(index = non_skin_index, inplace=True)
df_valid.groupby(["labels", "skin tone"]).size()
```

```python
non_skin_index = ca.check_no_skin_indetified(df_new_test)
df_test.drop(index = non_skin_index, inplace=True)
df_test.groupby(["labels", "skin tone"]).size()
```

# Measure Distance

```python
ms = MeasureSkin()
ms.select_baseline_skin(df_new_valid)
```

```python
df_train = ms.measure(df_new_train)
df_valid = ms.measure(df_new_valid)
df_test = ms.measure(df_new_test)
```

```python
df_train.to_csv(f"{root}/dataframe/df_train.csv", index=False)
df_valid.to_csv(f"{root}/dataframe/df_valid.csv", index=False)
df_test.to_csv(f"{root}/dataframe/df_test.csv", index=False)
```
